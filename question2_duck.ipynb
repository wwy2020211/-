```python
class FlyBehavior:                                     #飞
    ''' Interface class: FlyBehavior '''
    def fly(self):
        return

class FlyWithWing(FlyBehavior):               #翅膀飞行
    def fly(self):
        print("I am flying with wings!")
class FlyNoWay(FlyBehavior):                    #不会飞
    def fly(self):
        print("I cannot fly!")
class feather:                               #外表
    ''' Interface Behavior: QuackBehavior '''
    def color(self):
        return
class grey(feather):                 #灰色
    def color(self):
        print("my feather is grey!")
class grey_red(feather):                 #灰色和红色
    def color(self):
        print("my feather is grey and my head is red")
class white(feather):                          #白色
    def color(self):
        print("my feather is white")
class white_black(feather):                      #黑白
    def color(self):
        print("my feather is white and black")
class QuackBehavior:
    ''' Interface Behavior: QuackBehavior '''
    def quack(self):
        return
class Quack(QuackBehavior):              #叫声
    def quack(self):
        print("Quack!")
class Squeak(QuackBehavior):                #嘎嘎嘎
    def quack(self):
        print("嘎嘎嘎")
class MuteQuack(QuackBehavior):            #摩擦声
    def quack(self):
        print("橡皮与空气摩擦")
class quiet(QuackBehavior):                #不发声
    def quack(self):
        print("不发声")
class Duck:
    '''Base class: Duck. All ducks are inherent from this class'''
    def __init__(self, flyParam, quackParam,feather):
        self.flyBehavior = flyParam
        self.quackBehavior = quackParam
        self.feather=feather
    def performFly(self):
        self.flyBehavior.fly()
    def performQuack(self):
        self.quackBehavior.quack()
    def appearance(self):
        self.feather.color()
        return 
    def display(self):
        return
class RedDuck(Duck):
    def __init__(self, flyParam=FlyWithWing(), quackParam=Squeak(),feather=grey_red()):
        Duck.__init__(self, flyParam, quackParam,feather)
    def display(self):
        print("******************************************************")
        print("hello!!!I am a red duck!")
        return
class GreyDuck(Duck):
    def __init__(self, flyParam=FlyWithWing(), quackParam=Squeak(),feather=grey()):
        Duck.__init__(self, flyParam, quackParam,feather)
    def display(self):
        print("******************************************************")
        print("hello!!!I am a grey duck!")
class CottonDuck(Duck):
    def __init__(self, flyParam=FlyNoWay(), quackParam=quiet(),feather=white()):
        Duck.__init__(self, flyParam, quackParam,feather)
    def display(self):
        print("******************************************************")
        print("hello!!!I am a cotton duck!")
class RubberDuck(Duck):
    def __init__(self, flyParam=FlyNoWay(), quackParam=MuteQuack(),feather=white_black()):
        Duck.__init__(self, flyParam, quackParam,feather)
    def display(self):
        print("******************************************************")
        print("hello!!!I am a rubber duck!")
if __name__ == "__main__":
    duck = GreyDuck()
    duck.display()
    duck.performFly()
    duck.performQuack()
    duck.appearance()
       
    duck = RedDuck()
    duck.display()
    duck.performFly()
    duck.performQuack()
    duck.appearance()
    
    duck = CottonDuck()
    duck.display()
    duck.performFly()
    duck.performQuack()
    duck.appearance()
 
    duck = RubberDuck()
    duck.display()
    duck.performFly()
    duck.performQuack()
    duck.appearance()
```
